```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#UPLOADED DATASET1
```

```python
file_path="/content/dataset_1.xlsx"
df1=pd.read_excel(file_path)
df1
```

```python
#UPLOADED DATASET2
```

```python
file_path="/content/dataset_2.xlsx"
df2=pd.read_excel(file_path,index_col=0)
df2
```

```python
df2.columns
```

```python
df2.drop_duplicates
df2
```

```python
from matplotlib import pyplot as plt
df2['windspeed'].plot(kind='hist', bins=20, title='windspeed')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#MERGED DATASET1 AND DATASET2 WITH INNER JOIN
```

```python
merged_df = pd.merge(df1, df2, on='instant', how='inner')
merged_df
```

```python
merged_df.columns
```

```python
merged_df.drop_duplicates
merged_df
```

```python
#UPLOADED DATASET 3
```

```python
file_path="/content/dataset_3.xlsx"
df3=pd.read_excel(file_path)
df3
```

```python
df3.drop_duplicates
df3
```

```python
data_season = pd.concat([merged_df, df3])
data_season
```

```python
data_season=data_season.sort_values(by='instant')
data_season
```

```python
data_season.drop_duplicates
data_season
```

```python
from matplotlib import pyplot as plt
data_season['weekday'].plot(kind='hist', bins=20, title='weekday')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
